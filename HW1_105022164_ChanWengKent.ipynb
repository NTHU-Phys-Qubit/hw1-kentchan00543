Problem 1:Part 1

1D Random walk

We will concider two kinds of 1D random talk.

$dx$ is uniformly distributed between $[-1/2, +1/2]$.


For both kinds of the 1D random walk, you should do the following:

Plot a single trajectory.
Plot the histogram of the end point for an ensemble of trajectories.


```python
def my_random1(start,end,number):
    x= np.zeros(number);
    x_single = np.zeros(number);
    y_single = np.zeros(number);
    for p in range(0,number):
        x1 = 0
        y1 = 0
        for i in range(0,number):
            #to make sure dx and dy are within the range of limitation
            delta_x = np.random.random(number)*(end- start)+ start
            x1 = x1 + delta_x[i]
            x_single[i] = x1;
            #print(x1)
        x[p] = x1
    
        
        
    plt.figure()
    plt.title('Single Trajectory of Random Walk in 1 Dimensions', fontsize=20)
    plt.xlabel('Number of Steps', fontsize=20)
    plt.ylabel('x-coordinate', fontsize=20)
    plt.plot(x_single,'o', color='black');
    
    plt.figure()
    num_bins = 20
    n, bins, patches = plt.hist(x, num_bins, facecolor='blue', alpha=0.5)
    plt.xlabel('x-coordinate', fontsize=20)
    plt.ylabel('Numbers of Samples', fontsize=20)
    plt.title('Histogram of Random Walk( an ensemble of trajectories)—— x-component ', fontsize=20)
    plt.grid(True)
    plt.show() 
```

```python
my_random1(start=-1/2,end=+1/2,number=400)
```

Problem 1:Part 2

1D Random walk

$dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2.

For both kinds of the 1D random walk, you should do the following:

Plot a single trajectory.
Plot the histogram of the end point for an ensemble of trajectories.


```python
def my_random2(start,end,number):
    x= np.zeros(number);
    x_single = np.zeros(number);
    y_single = np.zeros(number);
    for p in range(0,number):
        x1 = 0
        for i in range(0,number):
            #to make sure dx is either -1 or 1 in half probabilities
            delta_x = np.random.choice(a=[start,end],size = number,p = [1/2,1/2]);
            x1 = x1 + delta_x[i]
            x_single[i] = x1;
        x[p] = x1
    
    plt.figure()
    plt.title('Single Trajectory of Random Walk in 1 Dimensions', fontsize=20)
    plt.xlabel('Number of Steps', fontsize=20)
    plt.ylabel('x-coordinate', fontsize=20)
    plt.plot(x_single,'o', color='black');
    
    plt.figure()
    num_bins = 20
    n, bins, patches = plt.hist(x, num_bins, facecolor='blue', alpha=0.5)
    plt.xlabel('x-coordinate', fontsize=20)
    plt.ylabel('Numbers of Samples', fontsize=20)
    plt.title('Histogram of Random Walk( an ensemble of trajectories)—— x-component ', fontsize=20)
    plt.grid(True)
    plt.show() 
    
    
    
```

Problem 2: Part 1

2D Random walk

We will concider three kinds of 2D random talk.

$dx$ and $dy$ are uniformly distributed between $[-1/2, +1/2]$.

For all kinds of the 2D random walk, you should do the following:

Plot a single trajectory.

Plot the distribution of the end point for an ensemble of trajectories.

Plot the histogram of the $r=x^2_f+y^2_f$ for an ensemble of trajectories.


```python
def my_random3(start,end,number):
    x= np.zeros(number);
    y = np.zeros(number);
    r = np.zeros(number);
    
    x_single = np.zeros(number);
    y_single = np.zeros(number);
    for p in range(0,number):
        x1 = 0
        y1 = 0
        for i in range(0,number):
            #to make sure dx and dy are within the range of limitation
            delta_x = np.random.random(number)*(end- start)+ start
            delta_y = np.random.random(number)*(end- start)+ start
            
            
            x1 = x1 + delta_x[i]
            y1 = y1 + delta_y[i]
            x_single[i] = x1;
            y_single[i] = y1;
            #print(x1)
        x[p] = x1
        y[p] = y1
        r[p] = math.sqrt(x[p]**2 + y[p]**2);
        #P[p] = (2*r[p]/number)*math.exp((-r[p]**2)/number);
        
    plt.figure()
    plt.title('Single Trajectory of Random Walk in 1 Dimensions', fontsize=20)
    plt.xlabel('Number of Steps', fontsize=20)
    plt.ylabel('x-coordinate', fontsize=20)
    plt.plot(x_single,'o', color='black');
    
    plt.figure()
    plt.title('Single Trajectory of Random Walk in 1 Dimensions', fontsize=20)
    plt.xlabel('Number of Steps', fontsize=20)
    plt.ylabel('y-coordinate', fontsize=20)
    plt.plot(y_single,'o', color='black');
    
    plt.figure()
    plt.title('Single Trajectory of Random Walk in 2 Dimensions', fontsize=20)
    plt.xlabel('x-coordinate', fontsize=20)
    plt.ylabel('y-coordinate', fontsize=20)
    plt.plot(x_single, y_single,'o', color='black');
    
    plt.figure()
    plt.title('Ensembles Trajectory of Random Walk in 2 Dimensions', fontsize=20)
    plt.xlabel('x-coordinate', fontsize=20)
    plt.ylabel('y-coordinate', fontsize=20)
    plt.plot(x, y,'o', color='black');
    
    plt.figure()
    num_bins = 20
    n, bins, patches = plt.hist(x, num_bins, facecolor='blue', alpha=0.5)
    plt.xlabel('x-coordinate', fontsize=20)
    plt.ylabel('Numbers of Samples', fontsize=20)
    plt.title('Histogram of Random Walk( an ensemble of trajectories)—— x-component ', fontsize=20)
    plt.grid(True)
    plt.show() 
    
    plt.figure()
    num_bins = 20
    n, bins, patches = plt.hist(y, num_bins, facecolor='blue', alpha=0.5)
    plt.xlabel('y-coordinate', fontsize=20)
    plt.ylabel('Numbers of Samples', fontsize=20)
    plt.title('Histogram of Random Walk( an ensemble of trajectories)—— y-component ', fontsize=20)
    plt.grid(True)
    plt.show()
    
    plt.figure()
    num_bins = 20
    n, bins, patches = plt.hist(r, num_bins, facecolor='blue', alpha=0.5)
    plt.xlabel('r:radius from the origin', fontsize=20)
    plt.ylabel('Numbers of Samples', fontsize=20)
    plt.title('Histogram of Random Walk( an ensemble of trajectories)—— r ', fontsize=20)
    plt.grid(True)
    plt.show() 
    
    
```

Problem 2: Part 2

2D Random walk

We will concider three kinds of 2D random talk.

Similarly for $dy$.


For all kinds of the 2D random walk, you should do the following:

Plot a single trajectory.

Plot the distribution of the end point for an ensemble of trajectories.

Plot the histogram of the $r=square(x^2_f+y^2_f)$ for an ensemble of trajectories.


```python
def my_random4(start,end,number):
    x= np.zeros(number);
    y= np.zeros(number);
    r = np.zeros(number);
    x_single = np.zeros(number);
    y_single = np.zeros(number);
    for p in range(0,number):
        x1 = 0
        y1 = 0;
        for i in range(0,number):
            #to make sure dx is either -1 or 1 in half probabilities
            delta_x = np.random.choice(a=[start,end],size = number,p = [1/2,1/2]);
            delta_y = np.random.choice(a=[start,end],size = number,p = [1/2,1/2]);
            x1 = x1 + delta_x[i];
            y1 = y1 + delta_y[i];
            x_single[i] = x1;
            y_single[i] = y1;
            
        x[p] = x1
        y[p] = y1
        r[p] = math.sqrt(x[p]**2 + y[p]**2);
        
    plt.figure()
    plt.title('Single Trajectory of Random Walk in 1 Dimensions', fontsize=20)
    plt.xlabel('Number of Steps', fontsize=20)
    plt.ylabel('x-coordinate', fontsize=20)
    plt.plot(x_single,'o', color='black');
    
    plt.figure()
    plt.title('Single Trajectory of Random Walk in 1 Dimensions', fontsize=20)
    plt.xlabel('Number of Steps', fontsize=20)
    plt.ylabel('y-coordinate', fontsize=20)
    plt.plot(y_single,'o', color='black');
    
    plt.figure()
    plt.title('Single Trajectory of Random Walk in 2 Dimensions', fontsize=20)
    plt.xlabel('x-coordinate', fontsize=20)
    plt.ylabel('y-coordinate', fontsize=20)
    plt.plot(x_single, y_single,'o', color='black');
    
    plt.figure()
    plt.title('Ensembles Trajectory of Random Walk in 2 Dimensions', fontsize=20)
    plt.xlabel('x-coordinate', fontsize=20)
    plt.ylabel('y-coordinate', fontsize=20)
    plt.plot(x, y,'o', color='black');
    
    plt.figure()
    num_bins = 20
    n, bins, patches = plt.hist(x, num_bins, facecolor='blue', alpha=0.5)
    plt.xlabel('x-coordinate', fontsize=20)
    plt.ylabel('Numbers of Samples', fontsize=20)
    plt.title('Histogram of Random Walk( an ensemble of trajectories)—— x-component ', fontsize=20)
    plt.grid(True)
    plt.show() 
    
    plt.figure()
    num_bins = 20
    n, bins, patches = plt.hist(y, num_bins, facecolor='blue', alpha=0.5)
    plt.xlabel('y-coordinate', fontsize=20)
    plt.ylabel('Numbers of Samples', fontsize=20)
    plt.title('Histogram of Random Walk( an ensemble of trajectories)—— y-component ', fontsize=20)
    plt.grid(True)
    plt.show() 
    
    plt.figure()
    num_bins = 20
    n, bins, patches = plt.hist(r, num_bins, facecolor='blue', alpha=0.5)
    plt.xlabel('r:radius from the origin', fontsize=20)
    plt.ylabel('Numbers of Samples', fontsize=20)
    plt.title('Histogram of Random Walk( an ensemble of trajectories)—— r ', fontsize=20)
    plt.grid(True)
    plt.show() 
```

Problem 2: Part 3

2D Random walk

We will concider three kinds of 2D random talk.

$(dx, dy)=(\cos(\theta), \sin(\theta))$ where $\theta$ is uniformly distributed between $[0, 2\pi]$.

For all kinds of the 2D random walk, you should do the following:

Plot a single trajectory.

Plot the distribution of the end point for an ensemble of trajectories.

Plot the histogram of the $r=x^2_f+y^2_f$ for an ensemble of trajectories.


```python
def my_random5(L,number):
    x= np.zeros(number);
    y = np.zeros(number);
    r = np.zeros(number);
    x_single = np.zeros(number);
    y_single = np.zeros(number);
    theta1 = np.zeros(number);
    
    for p in range(0,number):
        x1 = 0;
        y1 = 0;
        
        for i in range(0,number):
            theta = (np.random.random(1))*2*math.pi;
            theta1 = math.degrees(theta);
                
            #to make sure dx and dy are within the range of limitation
            delta_x = L *(math.cos(theta1));
            delta_y = L *(math.sin(theta1));
            x1 = x1 + delta_x
            y1 = y1 + delta_y
            x_single[i] = x1;
            y_single[i] = y1;
            #print(x1)
        x[p] = x1
        y[p] = y1
        r[p] = math.sqrt(x[p]**2 + y[p]**2);
        
    plt.figure()
    plt.title('Single Trajectory of Random Walk in 1 Dimensions', fontsize=20)
    plt.xlabel('Number of Steps', fontsize=20)
    plt.ylabel('x-coordinate', fontsize=20)
    plt.plot(x_single,'o', color='black');
    
    plt.figure()
    plt.title('Single Trajectory of Random Walk in 1 Dimensions', fontsize=20)
    plt.xlabel('Number of Steps', fontsize=20)
    plt.ylabel('y-coordinate', fontsize=20)
    plt.plot(y_single,'o', color='black');
    
    plt.figure()
    plt.title('Single Trajectory of Random Walk in 2 Dimensions', fontsize=20)
    plt.xlabel('x-coordinate', fontsize=20)
    plt.ylabel('y-coordinate', fontsize=20)
    plt.plot(x_single, y_single,'o', color='black');
    
    plt.figure()
    plt.title('Ensembles Trajectory of Random Walk in 2 Dimensions', fontsize=20)
    plt.xlabel('x-coordinate', fontsize=20)
    plt.ylabel('y-coordinate', fontsize=20)
    plt.plot(x, y,'o', color='black');
    
    plt.figure()
    num_bins = 20
    n, bins, patches = plt.hist(x, num_bins, facecolor='blue', alpha=0.5)
    plt.xlabel('x-coordinate', fontsize=20)
    plt.ylabel('Numbers of Samples', fontsize=20)
    plt.title('Histogram of Random Walk( an ensemble of trajectories)—— x-component ', fontsize=20)
    plt.grid(True)
    plt.show() 
    
    plt.figure()
    num_bins = 20
    n, bins, patches = plt.hist(y, num_bins, facecolor='blue', alpha=0.5)
    plt.xlabel('y-coordinate', fontsize=20)
    plt.ylabel('Numbers of Samples', fontsize=20)
    plt.title('Histogram of Random Walk( an ensemble of trajectories)—— y-component ', fontsize=20)
    plt.grid(True)
    plt.show()
    
    plt.figure()
    num_bins = 20
    n, bins, patches = plt.hist(r, num_bins, facecolor='blue', alpha=0.5)
    plt.xlabel('r:radius from the origin', fontsize=20)
    plt.ylabel('Numbers of Samples', fontsize=20)
    plt.title('Histogram of Random Walk( an ensemble of trajectories)—— r ', fontsize=20)
    plt.grid(True)
    plt.show() 
```
